```python
from retriever import retrieve, format_retrieved_documents
```

```python
collection_name = "ProjectDocuments"

retrieved_docs = retrieve(
    "cách quản lí project trong công ty?",
    collection_name,
    alpha=0.5,
    use_reranker=True,
    top_k=5
)
```

```python
retrieved_docs.objects[0].properties
```

```python
for i, obj in enumerate(retrieved_docs.objects):
    print(f'Document {i}:')
    print(f"Content: {obj.properties['content']}")  
    print(f"Distance: {obj.metadata.score}")
    print("-" * 80)
    
```

```python
print(format_retrieved_documents(retrieved_docs))
```
